```python
import numpy as np
#Política inicial del agente

def initial_policy(state):
  action=env.action_space.sample()
  while state[action+1]==1:
    action=env.action_space.sample()
  return action
#Generar episodios con la política inicial

def generate_episode(env,policy):
  episode=[]
  state=env.reset()
  recompensa=0
  while True:
    action=policy(state)
    next_state, reward, done,_=env.step(action)
    #recompensa=recompensa+reward
    episode.append((state,action,reward))
    state=next_state
    if done:
      break
  return episode

for i in range(10):
  print("\nEspisodio:", i)
  print(generate_episode(env,initial_policy))
    
```

```python

#Bucle principal
from collections import defaultdict
import sys

def mc_prediction_V(env, num_episodes, generate_episode, policy, gamma=1.0):

    returns_sum = defaultdict(float)
    N = defaultdict(int)
    V = defaultdict(float)
    
    for episode_number in range(1, num_episodes+1):
        
        if episode_number % 10000 == 0: 
            print("\repisode {}/{}.".format(episode_number, num_episodes), end="")
            sys.stdout.flush()
        
        episode = generate_episode(env, policy)
        states, actions, rewards = zip(*episode)
        
        for i, state in enumerate(states): #El probema está con el formato de state
            state_list=[]  #Conversión de list al formato adecuado 
            for j in state:
              state_list.append(j)
            state_tuple=tuple(state_list)
            returns_sum[state_tuple] += sum(r*gamma**k for k, r in enumerate(rewards[i:]))
            N[state_tuple] += 1.0
            V[state_tuple] = returns_sum[state_tuple] / N[state_tuple]
    return V
```

```python
NUM_EPISODES = 200000
GAMMA = 1
V = mc_prediction_V(env, NUM_EPISODES, generate_episode, initial_policy, GAMMA)
```

```python
#Bucle principal para Q
def mc_prediction_Q(env, num_episodes, generate_episode, policy, gamma=1.0):

    returns_sum = defaultdict(lambda: np.zeros(env.action_space.n))
    N = defaultdict(lambda: np.zeros(env.action_space.n))
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    
    for episode_number in range(1, num_episodes+1):
        
        if episode_number % 10000 == 0: # monitor progress
            print("\repisode {}/{}.".format(episode_number, num_episodes), end="")
            sys.stdout.flush()
        
        episode = generate_episode(env, policy)
        states, actions, rewards = zip(*episode)
        
        for i, state in enumerate(states):
            state_list=[]  #Conversión de list al formato adecuado 
            for j in state:
              state_list.append(j)
            state_tuple=tuple(state_list)
            returns_sum[state_tuple][actions[i]] += sum(r*gamma**k for k, r in enumerate(rewards[i:]))
            N[state_tuple][actions[i]] += 1
            Q[state_tuple][actions[i]] = returns_sum[state_tuple][actions[i]] / N[state_tuple][actions[i]]
    return Q
```

```python
def get_probs(Q_s, epsilon, nA):
    policy_s = np.ones(nA) * epsilon / nA
    max_action = np.argmax(Q_s)
    policy_s[max_action] = 1 - epsilon + (epsilon / nA)
    return policy_s
    
def generate_episode_from_Q(env, Q, epsilon, nA):
    episode = []
    state = env.reset()
    while True:
        state_list=[]
        for j in state:
          state_list.append(j)
          state_tuple=tuple(state_list)
        probs = get_probs(Q[state_tuple], epsilon, nA)
        action = np.random.choice(np.arange(nA), p=probs) if state_tuple in Q else env.action_space.sample()

        next_state, reward, done, _ = env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
    return episode


def update_Q(env, episode, Q, alpha, gamma):
    
    states, actions, rewards = zip(*episode)

    for i, state in enumerate(states):
        state_list=[]
        for j in state:          
            state_list.append(j)
            state_tuple=tuple(state_list)
        old_Q = Q[state_tuple][actions[i]] 
        Q[state_tuple][actions[i]] = old_Q + alpha*(sum(r*gamma**k for k, r in enumerate(rewards[i:])) - old_Q)
    return Q
```

```python


def MC_control(env, num_episodes, alpha, eps_decay, gamma):
    EPS_START = 1.0
    EPS_MIN = 0.05

    nA = env.action_space.n
    Q = defaultdict(lambda: np.zeros(nA))

    epsilon = EPS_START
    
    for episode in range(1, num_episodes+1):
        
        if episode % 1000 == 0: # monitor progress
            print("\rEpisode {}/{}.".format(episode, num_episodes), end="")
            sys.stdout.flush()

        epsilon = max(epsilon*eps_decay, EPS_MIN)
        episode_generated = generate_episode_from_Q(env, Q, epsilon, nA)
        Q = update_Q(env, episode_generated, Q, alpha, gamma)
    state_list=[]
    for j in state:
      state_list.append(j)
      state_tuple=tuple(state_list)    
    policy = dict((state_tuple,np.argmax(actions)) for state_tuple, actions in Q.items())
    return policy, Q

```
